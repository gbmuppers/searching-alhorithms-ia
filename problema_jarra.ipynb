En el problema de las jarras (A,B,C) se tienen dos jarras sin marcas de medición, una de A litros de capacidad y otra de B. También se dispone de una bomba que permite llenar las jarras de agua.
El problema de las jarras (A,B,C) consiste en determinar cómo se puede lograr tener exactamente C litros de agua en alguna de las dos jarras.

```python
def main():

    nodo_inicial = [[0, 0]]
    jarras = get_jarras()
    cantidad_objetivo = get_aguaf(jarras)
    check_dict = {}
    profundidad = get_boolean()
    search(nodo_inicial, jarras, cantidad_objetivo, check_dict, profundidad)

def get_index(node):

    return pow(7, node[0]) * pow(5, node[1])


def get_boolean():
    s=1
    return s==2

def get_jarras():

    jarras = []
    
    temp = int(input("Agua en la primera Jarra: "))
    while temp < 1:
        temp = int(input("No es valor valido: "))       
    jarras.append(temp)
    
    temp = int(input("Agua en la segunda Jarra: : "))
    while temp < 1:
        temp = int(input("No es valor valido: "))     
    jarras.append(temp)
    
    return jarras

def get_aguaf(jarras):

    max_amount = max(jarras[0], jarras[1])
    s = "Ingrese la cantidad de agua deseada (1 - {0}): ".format(max_amount)
    cantidad_objetivo = int(input(s))
    while cantidad_objetivo < 1 or cantidad_objetivo > max_amount:
        cantidad_objetivo = int(input("Ingrese un valor valido (1 - {0}): ".format(max_amount)))
        
    return cantidad_objetivo

def is_goal(path, cantidad_objetivo):

    return path[-1][0] == cantidad_objetivo or path[-1][1] == cantidad_objetivo

def been_there(node, check_dict):

    return check_dict.get(get_index(node), False)


def next_transitions(jarras, path, check_dict):
  
    result = []
    next_nodes = []
    node = []
    
    a_max = jarras[0]
    b_max = jarras[1]
    
    a = path[-1][0]  # cantidad inicial en la primera jarra
    b = path[-1][1]  # cantidad inicial en la segunda jarra

  
    node.append(a_max)
    node.append(b)
    if not been_there(node, check_dict):
        next_nodes.append(node)
    node = []

   
    node.append(a)
    node.append(b_max)
    if not been_there(node, check_dict):
        next_nodes.append(node)
    node = []

    
    node.append(min(a_max, a + b))
    node.append(b - (node[0] - a))  # b - ( a' - a)
    if not been_there(node, check_dict):
        next_nodes.append(node)
    node = []

   
    node.append(min(a + b, b_max))
    node.insert(0, a - (node[0] - b))
    if not been_there(node, check_dict):
        next_nodes.append(node)
    node = []

    
    node.append(0)
    node.append(b)
    if not been_there(node, check_dict):
        next_nodes.append(node)
    node = []

    
    node.append(a)
    node.append(0)
    if not been_there(node, check_dict):
        next_nodes.append(node)

   
    for i in range(0, len(next_nodes)):
        temp = list(path)
        temp.append(next_nodes[i])
        result.append(temp)


        

    return result


def transition(old, new, jarras):
    a = old[0]
    b = old[1]
    a_prime = new[0]
    b_prime = new[1]
    a_max = jarras[0]
    b_max = jarras[1]

    if a > a_prime:
        if b == b_prime:
            return "\t".format(a_max)
        else:
            return "\t".format(a_max, b_max)
    else:
        if b > b_prime:
            if a == a_prime:
                return "\t".format(b_max)
            else:
                return "\t".format(b_max, a_max)
        else:
            if a == a_prime:
                return "\t".format(b_max)
            else:
                return "\t".format(a_max)

"""
    Nos retorna la lista de nodos seguida para obtener el resultado imprime la ruta de la meta
"""
def print_path(path, jarras):
    print("Inicio:\n        ", path[0])
    for i in  range(0, len(path) - 1):
       print(i+1,":", transition(path[i], path[i+1], jarras), path[i+1])


def search(nodo_inicial, jarras, cantidad_objetivo, check_dict, profundidad):
    print("Problema de las jarras por búsqueda en profundidad \n")
    state_counter = 0
    goal = []
    accomplished = False
    q = collections.deque()
    q.appendleft(nodo_inicial)
    start_dfs_Jarras = time.time()
    while len(q) != 0:
        path = q.popleft()
        check_dict[get_index(path[-1])] = True
        if len(path) >= 2:
            state_counter = state_counter + 1
            transition(path[-2], path[-1], jarras)
        if is_goal(path, cantidad_objetivo):
            state_counter = state_counter + 1
            accomplished = True
            goal = path
            break

        next_moves = next_transitions(jarras, path, check_dict)
        for i in next_moves:
            if profundidad:
                q.appendleft(i)
            else:
                q.append(i)
    end_dfs_Jarras = time.time()
    if accomplished:
        print("\n\nTiempo en encontrar la solucion:",end_dfs_Jarras - start_dfs_Jarras,"segundos\n")
        print(state_counter+1," estados recorridos\n")
        print_path(goal, jarras)
    else:
        print("I failed you! :c")


if __name__ == '__main__':
    main()
```
